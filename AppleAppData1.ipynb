## **Great**

### **Reason for doing this ?**
- As we can fill the confusing values with the mean value of the numeric columns.
- For that, first we will get all the numeric values in one object.
- Next, we will impute that values with the mean of the rest of the values.

```python
df['Content_Rating'].value_counts()
```

- We will define a function that will replace the `Not yet rated` values with the  `NaN` values and can deal with them while dealing with the missing values.

```python
def convert_chars(Content_Rating):
  if isinstance(Content_Rating, str):
    if "Not yet rated" in Content_Rating:
      return np.nan
    return Content_Rating
```

```python
df['Content_Rating'] = df['Content_Rating'].apply(convert_chars)
```

## **Insight #6**
---
- For this insight we will compare columns `DeveloperId` , `Released` and `Updated`
- The purpose for doing this is that we can see how engaged a developer is with their apps that are being released
- Also to see the last time the app was updated

```python
# Convert 'Released' and 'Updated' columns to datetime
df['Released'] = pd.to_datetime(df['Released'])
df['Updated'] = pd.to_datetime(df['Updated'])
developer_engagement = df.groupby('DeveloperId').agg({
    'Released': 'min',
    'Updated': 'max'
})
developer_engagement['Time_to_Update'] = developer_engagement['Updated'] - developer_engagement['Released']

print(developer_engagement)
```

- Displaying the data for only top 5 developers

```python
top_5_developers=developer_engagement.head(5)
```

```python
top_5_developers.plot(kind='bar', y='Time_to_Update', figsize=(10, 6))
plt.xlabel('Developer ID')
plt.ylabel('Days')
plt.title('Time Difference between Release and Last Update by Developer')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
